```
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```
pokedex=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATA/pokedex_(Update_05.20).csv')
pokedex.head()
```

# lmplot()

```
sb.lmplot(x='attack',y='defense',data=pokedex)
```

```
sb.lmplot(x='attack',y='defense',data=pokedex,fit_reg=False,hue='generation',palette='magma')
```

# boxplot()

```
stats=pokedex[['hp',	'attack',	'defense',	'sp_attack',	'sp_defense',	'speed']]
stats.head()
```

```
sb.boxplot(data=stats)
```

# Pokemon color palette

```
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#705848',  # Dark
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#98D8D8',  # Ice
                    '#EE99AC',  # Fairy
                    '#B8B8D0',  # Steel
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost                   
                    '#7038F8',  # Dragon
                    '#A890F0',  # Flying
                   ]
```

# violinplot()

```
plt.figure(figsize=(15,5))
sb.violinplot(x='type_1',y='attack',data=pokedex,palette=pkmn_type_colors)
```

# stripplot()

```
plt.figure(figsize=(15,5))

sb.stripplot(x='type_1',y='attack',data=pokedex,palette=pkmn_type_colors)
```

# Swarmplot()

```
sub_df=pokedex[:151]
```

```
# 151 pokemons are plotted
plt.figure(figsize=(15,6))
sb.swarmplot(x='type_1',y='attack',data=sub_df,palette=pkmn_type_colors)
```

```
plt.figure(figsize=(15,6))

sb.violinplot(x='type_1',y='attack',data=sub_df,inner=None,palette=pkmn_type_colors)

sb.swarmplot(x='type_1',y='attack',data=sub_df,color='k',alpha=0.6)

plt.title('Attack vs type')
```

# heatmap()

```
corr=stats.corr()

sb.heatmap(corr)
```

# distplot()

```
sb.set_style('whitegrid')
sb.distplot(stats.attack,hist_kws={'edgecolor':'black'})
```

# countplot()

```
plt.figure(figsize=(15,6))

sb.countplot(x=pokedex.type_1,palette=pkmn_type_colors)
```

```
plt.figure(figsize=(15,5))
g=sb.catplot(x='type_1',y='attack',data=pokedex[:400],hue='generation',col='generation',kind='swarm')
g.set_xticklabels(rotation=-45)
```

# kdeplot()

```
sb.kdeplot(x=pokedex.sp_attack,y=pokedex.sp_defense)
```

```
sb.kdeplot(x=pokedex.sp_attack,y=pokedex.sp_defense)
sb.scatterplot(x=pokedex.sp_attack,y=pokedex.sp_defense,s=15)
```

# jointplot()

```

sb.jointplot(x=pokedex.attack,y=pokedex.defense)
```

```
sb.jointplot(x='sp_attack',y='sp_defense',data=pokedex,kind='hex')
```

# pairplot()

```
sb.pairplot(stats)
```

# rugplot()

```
sb.rugplot(pokedex.total_points)
```

```
sb.kdeplot(pokedex.hp)
```

# barplot()

```
plt.figure(figsize=(15,5))
sb.set_style('darkgrid')
sb.barplot(x='type_1',y='speed',data=pokedex,palette=pkmn_type_colors)
```

```
plt.figure(figsize=(15,5))
sb.set_style('darkgrid')
sb.barplot(x='type_1',y='speed',data=pokedex,palette=pkmn_type_colors,estimator=np.std)
```

# clustermap

```
flights=sb.load_dataset('flights')
flights.head()
```

```
flights.pivot_table(index='month', columns='year', values='passengers')
# index is your row, columns and values are numbers stored at the intersection of x,y row and column.
```

```
fp = flights.pivot_table(index='month',columns='year',values='passengers')
sb.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
#  cmap schemes : magma, coolwarm, spring
```

```
sb.clustermap(fp,cmap='coolwarm')
```

```
sb.clustermap(fp,cmap='coolwarm',standard_scale=1)
```
